```python
from travelling_salesman import Route, map24, map150, map1000, map24_random, map150_greedy, map150_random, calculate_distance, create_initial_greedy
from matplotlib import pyplot as plt

map24.plot()
map24_random.plot()
```

```python
map150.plot()
map150_random.plot()
```

```python
map150_greedy.plot()
```

```python
%%time
dists = []
smallest = float("inf")
smallest_path = 0
smallest_index = 0

for i in range(len(map150.coordinates)):
    p = create_initial_greedy(map150, start=i)
    d = calculate_distance(map150, p)
    dists.append(d)
    if d < smallest:
        smallest = d
        smallest_path = p
        smallest_index = i

plt.plot(dists)
plt.xlabel("Starting index")
plt.ylabel("Distance")
plt.show()

best_greedy = Route(map150, smallest_path)
print(f"Best greedy route. Starting index = {smallest_index}")
best_greedy.plot()
```
